### 在数据分析之前，先明确变量的度量类型（名义、等级、连续）
1. 名义变量可以是字符，也可以是数字，名义变量没有顺序
2. 等级变量相较于名义变量有顺序之分，比如小学，中学，大学
3. 连续变量必须是数值型
4. 如果我们关心两个等级之间的差值，可以作为连续变量，如果我们不关心两个等级之间差值（小学中学转化为1,2，我们不会关心差值），会作为等级变量（WOE转化，将变量分组）--因为连续变量容易产生noise

### 连续分布形式
1. 正态分布（自然界）
2. 泊松分布（网页点击量，队伍排队）
3. 伽马分布（灾难对我们的损失，理赔）
4. 对数正态分布（人的收入情况，企业利润情况，人的工资）右偏最严重的<br />描述性统计：直接取中位数；建模预测，对原始数据取对数，就是正态分布了<br />
一般只取1和4

二. 离中趋势（分散程度）<br />
1. 极差、四分位差（上分位点和下分位点之差）
2. 平均绝对偏差
3. 方差和标准差: <br />除以n-1,指的是自由度，公式中平均值相当于用了一份信息<p>

### 为什么使用IQR，不适用标准差？<br />因为标准差是基于数据符合正态分布，如果是左偏或右偏，就不适用了。（特别的95%范围内，即使正态分布的理论）

三. 偏态和峰态（形状）<p>
正态分布： Skewness = 0 左偏负值，右偏正值
